```python
fc_image = FileChooser('.')
fc_image.title = '<b>Select image to denoise (.tif)</b>'
fc_image.filter_pattern = ['*.tif', '*.tiff']

fc_model = FileChooser('.')
fc_model.title = '<b>Select trained model (eg: k3d2f64_0001)</b>'
fc_model.show_only_dirs = True

display(widgets.VBox([
    widgets.HTML("<h3>Loading data and model</h3>"),
    fc_model,
    fc_image
]))
```

```python
def verifier_format_modele(nom_modele):
    pattern = r"k\d+.*d\d+"
    return bool(re.search(pattern, nom_modele, re.IGNORECASE))

if not fc_image.selected or not fc_model.selected:
    raise ValueError("Error : Select a model and an image first.")

image_path = Path(fc_image.selected)
output_path = image_path.with_name(f"{image_path.stem}_denoised_care{image_path.suffix}")

model_full_path = Path(fc_model.selected)
model_name = model_full_path.name       # eg: training_k3d1l64_0001
model_dir = model_full_path.parent      # eg: /Users/.../test_data/models

if not verifier_format_modele(model_name):
    print(f"Warning : Model name '{model_name}' is not in the expected format.")

print(f"Image: {image_path}")
print(f"Saving name: {output_path}")
print(f"Model used: {model_name}")
print("Starting...")

start = time.time()

do_denoising_3D(
    input_path=str(image_path), 
    output_path=str(output_path), 
    model_name=model_name,
    basedir=str(model_dir) 
)

end = time.time()
print(f"Finished in {(end - start) / 60:.2f} minutes !")
```
